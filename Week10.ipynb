```python
#pip install -U scikit-learn
from sklearn.datasets import fetch_california_housing
data = fetch_california_housing()

print(data.DESCR)
```

**Exploring the Data with Pandas**

```python
import pandas as pd

df = pd.DataFrame(data.data, columns=data.feature_names) #create a dataframee from the 'data' object
pd.set_option('display.precision', 2)
pd.set_option('display.max_columns', 9)
pd.set_option('display.width', None)
df.head()
```

**Visualization**

```python
import matplotlib.pyplot as plt
import seaborn as sns


# Sample 10% of the data for visualization
df_sample = df.sample(frac=0.1, random_state=17)

#get dataframe of MedHouseVal
target_data = data.target
df_target = pd.DataFrame(target_data, columns=data.target_names)

# Sample 10% of the target for visualization
target_sample = df_target.sample(n=len(df_sample), random_state=17)

df_sample['MedHouseVal'] = target_sample

sns.set(font_scale=2)
sns.set_style('whitegrid')

for feature in df_sample.columns:
  plt.figure(figsize=(16,9))
  sns.scatterplot(data=df_sample, x=feature, y='MedHouseVal', hue='MedHouseVal', palette='cool', legend=False)
```

**Conclusions:**

**Graph 1:** From the scatter plot in ghraph 1, it appears that there is a positive correlation between MedInc and MedHouseVal. This means that as the median income in a block increases, the median house value in that block also tends to increase. The points are more densely packed at lower values of MedInc and MedHouseVal, indicating that there are more blocks with lower median income and house values. As the values increase, the points become more spread out, indicating fewer blocks with high median income and house values.
