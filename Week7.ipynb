 **Showing Plots**

```python
import matplotlib as mp
import matplotlib.pyplot as plt
import numpy as np

plt.style.use('classic')

x = np.linspace(0, 10, 100)
print(x)

plt.plot(x, np.sin(x))
plt.plot(x, np.cos(x))

plt.show()
```

**Drawing a Line**

```python
xpoints= np.array([0, 10])
ypoints = np.array([0, 20])

plt.plot(xpoints, ypoints)
plt.show
```

To plot just the markers, we can use the parameter o

**The `markersize` or `ms` keyword argument
can be used to set the size of the marker on
a line.**


```python
xpoints = np.array([1, 5, 11, 12, 17, 21])
ypoints = np.array([2, 8, 11, 18, 25, 45])

plt.plot(xpoints, ypoints, 'Dg', markersize = 20)
plt.show()
```

             Marker Color(s)

**The `markerfacecolor` or `mfc` and `markeredgecolor` or `mec`
keywords can be used to set colors of the marker and its
edge.**

```python
xpoints = np.array([1, 5, 11, 12, 17, 21])
ypoints = np.array([2, 8, 11, 18, 25, 45])

plt.plot(xpoints, ypoints, 'D--g', ms = 15, mfc = 'b', mec = 'r')
plt.show()
```

                  Plot Labels

**The xlabel() and ylabel() functions can be used to set the
labels of x and y-axes.**

```python
xpoints = np.array([1, 5, 11, 12, 17, 21])
ypoints = np.array([2, 8, 11, 18, 25, 45])

plt.plot(xpoints, ypoints, 'D--g', ms = 15, mfc = 'b', mec = 'r')
plt.xlabel("x axis label")
plt.ylabel("y axis label")
plt.show()
```

                   Plot Title

**The `title()` function can be used to set the title of the plot.**

```python
xpoints = np.array([1, 5, 11, 12, 17, 21])
ypoints = np.array([2, 8, 11, 18, 25, 45])

plt.plot(xpoints, ypoints, 'D--g', ms = 15, mfc = 'b', mec = 'r')
plt.xlabel("x axis label")
plt.ylabel("y axis label")
plt.title("Out Plot Title")
plt.show()
```

  **Font Properties for Title & Labels**

```python
xpoints = np.array([1, 5, 11, 12, 17, 21])
ypoints = np.array([2, 8, 11, 18, 25, 45])

plt.plot(xpoints, ypoints, 'D--g', ms = 15, mfc = 'b', mec = 'r')
font1 = {'family' : 'serif', 'color': 'b', 'size' : 15}
font2 = {'family' : 'serif', 'color' : 'r', 'size': 20}
plt.xlabel("x axis label", font1)
plt.ylabel("y axis label", font1)
plt.title("Out Plot Title", font2)
plt.show()
```


     Color Maps
A colormap is like a list of colors. Matplotlib has several colormaps. Each color in
Matplotlib has a value that ranges from 0 to 100.
Consider the example of the `Viridis` colormap, its value 0 is the purple color and
100 has color yellow.

```python
x = np.array([5, 7, 8, 7, 2, 17, 2, 9, 4, 11, 12, 9, 6])
y = np.array([99.86, 87.88, 111.8, 86.1083, 87.94, 78, 77, 85, 86, 12, 454, 65, 55])
colors = np.array([0, 10, 20, 30, 40, 45, 50, 55, 60, 70, 80, 90, 100])

plt.scatter(x, y, c = colors, cmap= 'viridis')

plt.show()
```


       Showing the Color Map Bar

```python
x = np.array([5, 7, 8, 7, 2, 17, 2, 9, 4, 11, 12, 9, 6])
y = np.array([99.86, 87.88, 111.8, 86.1083, 87.94, 78, 77, 85, 86, 12, 454, 65, 55])
colors = np.array([0, 10, 20, 30, 40, 45, 50, 55, 60, 70, 80, 90, 100])

plt.scatter(x, y, c = colors, cmap= 'gist_rainbow')

plt.colorbar()

plt.show()
```


      Dot Size in Scatter Plot

```python
x = np.array([5, 7, 8, 7, 2, 17, 2, 9, 4, 11, 12, 9, 6])
y = np.array([99.86, 87.88, 111.8, 86.1083, 87.94, 78, 77, 85, 86, 12, 454, 65, 55])
sizes = np.array([50, 20, 100, 200, 400, 405, 500, 55, 60, 70, 800, 90, 100])

plt.scatter(x, y, c = colors, s= sizes)

plt.show()
```



        Transparency of Dots

```python
x = np.array([5, 7, 8, 7, 2, 17, 2, 9, 4, 11, 12, 9, 6])
y = np.array([99.86, 87.88, 111.8, 86.1083, 87.94, 78, 77, 85, 86, 12, 454, 65, 55])
sizes = np.array([50, 20, 100, 200, 400, 405, 500, 55, 60, 70, 800, 90, 100])

plt.scatter(x, y, c = colors, s= sizes, alpha=0.4)

plt.show()
```
