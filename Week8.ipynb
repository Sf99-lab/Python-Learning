> **Histogram(s) in Matplotlib**

```python
import numpy as np

np.random.seed(100)
x = np.random.normal(0, 10, 100)

print(x)
```

```python
styles = plt.style.available

for style in styles:
    plt.style.use(style)
    plt.hist(x)
    plt.show()
```

```python
styles = plt.style.available

for style in styles:
  plt.style.use(style)
  plt.hist(x, color = 'g')
  plt.show()
```

```python
styles = plt.style.available

for style in styles:
  plt.style.use(style)
  plt.hist(x, color= 'yellow')
  plt.show()
```

```python
styles = plt.style.available

for style in styles:
  plt.style.use(style)
  plt.hist(x, color = 'g', bins= 8)
  plt.show()
```

```python
plt.hist(x, color = 'yellow', bins= 5)
plt.show()
```

```python
styles = plt.style.available

for style in styles:
  plt.style.use(style)
  plt.hist(x, color = 'pink', bins= 8)
  plt.xlabel("x-axis")
  plt.ylabel("y-axis")
  plt.title("Out Histogram")
  plt.show()
```

```python
styles = plt.style.available

for style in styles:
  plt.style.use(style)
  plt.hist(x)
  plt.show()
```

```python
styles = plt.style.available

for style in styles:
  plt.style.use(style)
  plt.hist(x)
  plt.show()
```

```python
styles = plt.style.available

for style in styles:
  plt.style.use(style)
  plt.style.use('ggplot')
  plt.hist(x, rwidth = 0.3)
  plt.show()
```

```python
styles = plt.style.available

for style in styles:
  plt.style.use(style)
  plt.hist(x, rwidth = 0.6)
  plt.show()
```

```python
styles = plt.style.available

for style in styles:
  plt.style.use(style)
  plt.hist(x, histtype= 'step')
  plt.show()
```

```python
plt.hist(x, histtype= 'stepfilled')
plt.show()
```

```python
styles = plt.style.available

for style in styles:
  plt.style.use(style)
  plt.hist(x, histtype = 'barstacked')
  plt.savefig('histo.png')
  plt.savefig('histo.jpg')
  plt.savefig('histo.pdf')
  plt.show()
```

> **Pie Chart(s) in Matplotlib**

```python
import matplotlib.pyplot as plt
import numpy as np

x = np.array([ 55, 24, 12, 9])

plt.pie(x)
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

x = np.array([ 55, 24, 12, 9])
provLabels = [ "Punhjab", "KPK", "Sindh", "Balochistan"]

plt.pie(x, labels = provLabels)

plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

x = np.array([ 55, 24, 12, 9])
provLabels = [ "Punhjab", "KPK", "Sindh", "Balochistan"]

plt.pie(x, labels = provLabels, startangle = 90)

plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

x = np.array([ 55, 24, 12, 9])
provLabels = [ "Punhjab", "KPK", "Sindh", "Balochistan"]
myexp = [ 0.2, 0,0,0]

plt.pie(x, labels = provLabels, startangle = 0, explode = myexp)

plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

x = np.array([ 55, 24, 12, 9])
provLabels = [ "Punhjab", "KPK", "Sindh", "Balochistan"]
myexp = [ 0.2, 0,0,0]

plt.pie(x, labels = provLabels, startangle = 0, explode = myexp, shadow = True)

plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

x = np.array([ 55, 24, 12, 9])
provLabels = [ "Punhjab", "KPK", "Sindh", "Balochistan"]
wColors = ["g", "red", "b", "#4CAF50"]

plt.pie(x, labels = provLabels, colors = wColors)

plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

x = np.array([ 55, 24, 12, 9])
provLabels = [ "Punhjab", "KPK", "Sindh", "Balochistan"]
wColors = [ "g", "red", "b", "#4CAF50"]


plt.pie(x, labels = provLabels, colors = wColors)
plt.legend()

plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

x = np.array([ 55, 24, 12, 9])
provLabels = [ "Punhjab", "KPK", "Sindh", "Balochistan"]
wColors = [ "g", "red", "b", "#4CAF50"]


plt.pie(x, labels = provLabels, colors = wColors)
plt.legend(title = "Provinces")

plt.show()
```
