# **Reading and Writing Files / Exception Handling**
Create two text files by hand with in which sequences of ten numbers are written on each line. The sequence of numbers should be ended with a ".".  The numbers can be as under:

4 5 900 85 12 15 656 76 -90 155.
-90 12 15 0 899 36 10 111 -112 50.

and so on 10 different lines

Save both these in the "contents" folder of your Google CoLab as numbers1.txt and numbers2.txt.

Now write a program that reads a line from both these files one by one and then stores the product of numbers in both files in a file called products.txt and the division of numbers in numbers1.txt divided by those in numbers2.txt in a file called division.txt.

Your program must check for possible exceptions like empty file, non-existent file etc.

```python
try:
  with open("numbers1.txt", "r") as file:
    content_1 = file.read()

    numbers_as_strings1 = content_1.split()
    numeric_values_1 = [eval(num) for num in numbers_as_strings1]



  product_1 = 1
  for num in numeric_values_1:
    product_1 *= num

  print(product_1)

  with open("numbers2.txt", "r") as file2:
    content_2 = file2.read()

  numbers_as_strings2 = content_2.split()
  numeric_values_2 = [eval(num) for num in numbers_as_strings2]

  product_2 = 1
  for num in numeric_values_2:
    product_2 *= num

  print(product_2)


  with open('products.txt', 'w') as file:
    file.write(f"Products of numbers1.txt is: ${product_1}")
    file.write(f"\nProducts of numbers2.txt is: ${product_2}")

  division = []
  for i in range(min(len(numeric_values_1), len(numeric_values_2))):
    if numeric_values_2[i] != 0:
      division.append(round(numeric_values_1[i] / numeric_values_2[i], 3))
    else:
      division.append(float('inf'))
  print(division)
  with open('division.txt', 'w') as divfile:
    divfile.write("The division of numbers1.txt with numbers2.txt is as follow:\n"\
                   f"${division}")


except FileNotFoundError:
  print("File not found")
except ValueError:
  print("Not a valid value")
except IOError:
  print("unable to read file")
except TypeError:
  print("You are dealing oject of inappropriate type")
except NameError:
  print("a variable or function hasn't been defined yet")
except ZeroDivisionError:
  print('Attempting to divide by zero')
finally:
  if file:
    file.close()
  if file2:
    file2.close()
  if divfile:
    divfile.close()
```

Add methods for computing the area and perimeter of 2D shapes and surface area and volume of 3D shapes.

We may use these implemented classes to draw these shapes in a futre lab.

```python
import math

try:
    class Shape:
        def __init__(self, *args):
            pass

        def area(self):
            pass

        def perimeter(self):
            pass

    class TwoDimensionalShape(Shape):
        def __init__(self, *args):
            super().__init__(*args)

    class ThreeDimensionalShape(Shape):
        def __init__(self, *args):
            super().__init__(*args)

    class Circle(TwoDimensionalShape):
        def __init__(self, radius):
            super().__init__(radius)
            self.radius = radius

        def area(self):
            return math.pi * self.radius**2

        def perimeter(self):
            return 2 * math.pi * self.radius

    class Square(TwoDimensionalShape):
        def __init__(self, side_length):
            super().__init__(side_length)
            self.side_length = side_length

        def area(self):
            return self.side_length**2

        def perimeter(self):
            return 4 * self.side_length

    class Triangle(TwoDimensionalShape):
        def __init__(self, side1, side2, side3):
            super().__init__(side1, side2, side3)
            self.side1 = side1
            self.side2 = side2
            self.side3 = side3

        def area(self):
            # calculate area of a triangle
            s = (self.side1 + self.side2 + self.side3) / 2
            return math.sqrt(s * (s - self.side1) * (s - self.side2) * (s - self.side3))

        def perimeter(self):
            return self.side1 + self.side2 + self.side3

    class Sphere(ThreeDimensionalShape):
        def __init__(self, radius):
            super().__init__(radius)
            self.radius = radius

        def surface_area(self):
            return 4 * math.pi * self.radius**2

        def volume(self):
            return (4/3) * math.pi * self.radius**3

    class Cube(ThreeDimensionalShape):
        def __init__(self, side_length):
            super().__init__(side_length)
            self.side_length = side_length

        def surface_area(self):
            return 6 * self.side_length**2

        def volume(self):
            return self.side_length**3

    class Tetrahedron(ThreeDimensionalShape):
        def __init__(self, side_length):
            super().__init__(side_length)
            self.side_length = side_length

        def surface_area(self):
            return math.sqrt(3) * self.side_length**2

        def volume(self):
            return (self.side_length**3) / (6 * math.sqrt(2))

    # Example usage
    circle = Circle(5)
    print("Circle Area:", circle.area())
    print("Circle Perimeter:", circle.perimeter())

    square = Square(4)
    print("Square Area:", square.area())
    print("Square Perimeter:", square.perimeter())

    triangle = Triangle(3, 4, 5)
    print("Triangle Area:", triangle.area())
    print("Triangle Perimeter:", triangle.perimeter())

    sphere = Sphere(5)
    print("Sphere Surface Area:", sphere.surface_area())
    print("Sphere Volume:", sphere.volume())

    cube = Cube(4)
    print("Cube Surface Area:", cube.surface_area())
    print("Cube Volume:", cube.volume())

    tetrahedron = Tetrahedron(3)
    print("Tetrahedron Surface Area:", tetrahedron.surface_area())
    print("Tetrahedron Volume:", tetrahedron.volume())

except Exception as e:
    print("An error occurred:", e)
```
