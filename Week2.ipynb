# **Splitting and Joining Strings**
Take help from you lecture notes and google to find the difference between, ***split()***, ***rsplit()*** and ***join()*** methods for strings in Python. Produce an example of each in the following cell.

```python
original_string = "We can split Strings in Python"

split_string = original_string.split()
print("Here we go:", split_string)

rsplit_string = original_string.rsplit()
print("Here we go with rsplit() function: ", rsplit_string)

joined_string_with_delimiter = ' '.join(rsplit_string)
print("join() with delimiter ' ':", joined_string_with_delimiter)
```

The split() method is used to break a given string
into substrings based on a specified separator (delimiter).
Syntax: my_string.split(sep, maxsplit)



The key difference is that rsplit() starts splitting from the right end of the string. The syntax will be same.



The join() method is used to concatenate a list of strings into a single string.
It takes an iterable (e.g., a list) of strings and joins them using the specified separator.

Syntax: separator.join(iterable)
separator: The string that separates the elements in the iterable.
iterable: A list of strings you want to join.




# **Slicing**

Consider the following string:

"A quick brown fox jumps over the lazy dog."

Use the syntax for ***slicing*** in Python to produce slices of ***first four characters***, ***last four characters*** and the ***middle four characters*** from the above string in code cell below.

```python
string = "A quick brown fox jumps over the lazy dog."
first_four = string[:4] #For first 4four character
last_four = string[-4:] #last four character
middle_four = string[7:11] #middle four character
print(f"First four characters: {first_four}")
print(f"Last four characters: {last_four}")
print(f"Middle four characters: {middle_four}")
```

# **Strides**

Continue with the string in the previous cell print the following in the next cell.
1. Now you print the string obtained by taking strides of 2 from the beginning
2. Strides of 4 starting from the 10th character till the end
3. Try printing using negative indices with strides of 2 till index -10




```python
string = "A quick brown fox jumps over the lazy dog."
string_strides_2 = string[::2]
print(f"Strides of 2 from the beginning: {string_strides_2}")

from_10th_character_strides_4 = string[9::4]
print(f"Strides of 4 from the 10th character: {from_10th_character_strides_4}")

negative_indices_strides_2 = string[:-10:-2]
print(f"Negative indices with strides of 2: {negative_indices_strides_2}")
```

```python

string = "A quick brown fox jumps over the lazy dog"
split_result = string.split()
print("Split Output:", split_result)

list_a = [1, 2, 3]
list_b = list_a
list_a.append(4)
print("List A:", list_a)
print("List B (Aliased):", list_b)

list_c = [5, 6, 7]
list_d = list_c.copy()  # Cloning(In this case list_ and list_d are seprate object when we modify list_c its can't affect on list_d)
list_c.append(8)
print("List C:", list_c)
print("List D (Cloned):", list_d)
```
