Use the relational operators (<, <=. >, >=, ==, !=) between sets A = {1, 3, 5, 7, 9} and B = {3, 5, 9} and print the **comparison** results to screen. Then check the containment relationship betweeen A contained in B and B contained in A using the issubset() methods provided by Python. Show the output to screen for this operation also.

```python
A = {1, 3, 5, 7, 9}
B = {3, 5, 9}

#Comparison using relational operator
print("A < B:", A < B)
print("A <= B:", A <= B )
print("A > B:", A > B )
print("A >= B:", A >= B)
print("A==B:", A == B)
print("A!=B:", A != B)

#Containment Relationship check
print("\nA contained in B:", A.issubset(B))
print("B contained in A:", B.issubset(A))
```

Now use the mathematical set operations **union**, **intersection**, **difference**, **symmetric difference** and **disjoint** (using the isdisjoint() function) on sets A = {1, 3, 5, 7, 9, 11} and B = {2, 3, 4, 5, 7, 9}. Print the result of each of the mentioned operation on screen.

Once the above is completed also check the effect of add(), remove() and pop() methods on set A given above. You may have to provide some parameters for these methods. Write the effect of invoking each of these methods on the set.

Now re-write the code in ***Nested Loops*** section above in such a way that the aesterisks traingles in part a, b, c and d are so combined in one program that it looks a diamond of aesterisks. For example, one strategy could be to place d and a side by side and then print c and b immediately below them.

This can be done by defining functions for each of the above four parts above and then calling these functions in a proper way to show the desired output.

```python
def part_a():
  for i in range(1,10):
    for j in range(i):
      print("*", end='')
    print()

def part_b():
  for i in range(10):
    for j in range(10-i):
        print("*", end='')
    print()

def part_c():
  for i in range(10):
    for k in range(0,i):
      print('', end = ' ')
    for j in range(10-i):
      print("*", end='')
    print()

def part_d():
  for i in range(10):
    for k in range(10-i):
      print('', end = ' ')
    for j in range(i):
      print("*", end='')
    print()

def print_diamond():
  part_d()
  print("", end="")
  part_a()

  part_c()
  part_b()

# Print the diamond
print_diamond()
```

**Random Number Generation in Python**

If ***randrange()*** truly produces integers at random then every number in its range has an equal probability (or chance or likelihood) of being returned each time we call it. To show that the die faces 16 occur with equal likelihood, write Python code that simulates 6,000,000 die rolls. Meaning that your randrange() will be executed 6,000,000 times and you will record the occurences of each die face value in the corresponding variables (these six variables will be incremented when the corresponding face value appears/generated by the randrange()). When you run the program, each die face should occur approximately 1,000,000 times. If that happens it would mean that Python's randrange() do exhibit some sort of "***randomness***".

```python
import random

# Define variables to store counts for each die face
face_1 = 0
face_2 = 0
face_3 = 0
face_4 = 0
face_5 = 0
face_6 = 0

# Number of die rolls to simulate
num_rolls = 6_000_000

# Simulate die rolls and count occurrences
for _ in range(num_rolls):
  roll = random.randrange(1, 7)  # Generate random number between 1 and 6 (inclusive)
  if roll == 1:
    face_1 += 1
  elif roll == 2:
    face_2 += 1
  elif roll == 3:
    face_3 += 1
  elif roll == 4:
    face_4 += 1
  elif roll == 5:
    face_5 += 1
  else:
    face_6 += 1

# Print the results
print("Die Face\tOccurrences (Expected: 1,000,000)")
print("-------\t---------------")
print(f"1:\t\t{face_1}")
print(f"2:\t\t{face_2}")
print(f"3:\t\t{face_3}")
print(f"4:\t\t{face_4}")
print(f"5:\t\t{face_5}")
print(f"6:\t\t{face_6}")


```

*** This result shows some sort of randomness in Python's randsrange()***
